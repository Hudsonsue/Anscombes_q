## Section two - background of the dataset 

A paper, Graphs in Statistical Analysis [1], was published in 1973 by Francis J. Anscombe. 
This paper explored the reliance on numerical calculations rather than graphs and showed that four datasets
with almost identical variances, correlations and regression lines could in fact be completely dissimilar when plotted.

The summary statistics would imply that the four sets of data are almost identical yet when plotted it is evident that they vary to a large extent, this highlights the need when looking at data to examine data graphically before assuming any particular relationship. I will in this notebook import the dataset, prodce summary statistics and then graphs to illustrate Anscombe's quartet. 


It is unknown how he created the datasets [2]but several methods to develop similar datasets have been developed since.

## Section Three - Plot interesting aspects of the dataset.

Using seaborn I will below load in the dataset, display the dataset, produce and display summary statistics and plot the four datasets.

```python
# load the dataset#
data = sns.load_dataset("anscombe") 
data
```

```python
#produce summmary statistics#
data.groupby("dataset").describe()
```

```python
#plot graphs of the datasets using seaborn plots# 
sns.lmplot(x="x", y="y", col="dataset", hue="dataset", data=data,
           col_wrap=2, ci=None, palette="bright", height=5)
          
plt.show()
```
