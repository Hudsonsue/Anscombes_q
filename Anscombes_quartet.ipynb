## Section two - background of the dataset 

[Francis J. Anscombe](https://en.wikipedia.org/wiki/Frank_Anscombe)(13 May 1918 â€“ 17 October 2001) was an English statistician. He moved to the USA in the mid 1950s and was the founding chair of the Statistics Department at [Yale University](https://www.yale.edu/). In 1973 he published a paper, *Graphs in Statistical Analysis* [1].

This paper explored the reliance on numerical calculations rather than graphs and showed that four datasets
with almost identical variances, correlations and regression lines could in fact be completely dissimilar when plotted. It also looked at the effect of outliers on data. 

The summary statistics would imply that the four sets of data are almost identical yet when plotted it is evident that they vary to a large extent, this highlights the need when looking at data to examine data graphically before assuming any particular relationship. I will in this notebook import the dataset, produce summary statistics and then graphs to illustrate Anscombe's quartet. 

It is unknown how he created the datasets [2] but several methods to develop similar datasets have been developed since.

I used two methods of importing and plotting. Initially I imported the dataset and plotted using the seaborn package. From my reaserch for this assignment seaborn tutorials [3] suggests that seaborn can "*make exploring a dataset through visualization quick and easy, as doing so is just as (if not more) important than exploring a dataset through tables of statistics*", this appeared to me to be an ideal tool to demonstrate what Anscombe had outlined in his 1973 paper. 
I then used Pandas to read in a csv file [4], produce summary statistics using python and plotted with regression lines using mathplotlib. 

## Section three - plot interesting aspects of the dataset & Section four - calculate descriptive statistics of the variables in the dataset

### please note that I have combined the plots and descriptive statistics sections

Using seaborn I will below load in the dataset, display the dataset, produce and display summary statistics and plot the data.

```python
# load the dataset available in seaborn

data = sns.load_dataset("anscombe") 
#display the dataset

data
```

```python
#produce summmary statistics grouped by dataset

data.groupby("dataset").describe()
```

```python
# plot graphs of the datasets using seaborn plots
# include regression lines and two plots per pagewidth

sns.lmplot(x="x", y="y", col="dataset", hue="dataset", data=data,
           col_wrap=2, ci=None, palette="bright", height=5)
          
plt.show()
```

To highlight how easy the steps above were I will now use pandas to read in the csv file and display, them use the describe command to produce summary statistics and finally plot using Mathplotlib.

```python
## reading in the anscombe.csv file from column position 1 and displaying the dataset

df = pd.read_csv("anscombe.csv").iloc[:, 1:]
df
```

```python
# describe the summary statistics of the dataset 
df.describe()
```

```python
x = a =df['x1'].values
x4 =a = df['x4'].values
y1 =a = df['y1'].values
y2=a = df['y2'].values
y3 =a = df['y3'].values
y4 =a = df['y4'].values
```

```python
# calculate the correlation coefficients for all four datasets
# as X1, X2 and X3 are identical I am calling X and X4
# as the command returns a matrix I am using [1,0] to display the first array position only
print("Correlation coefficient dataset 1  ", np.corrcoef(x, y1)[1,0])
print("Correlation coefficient dataset 2  ", np.corrcoef(x, y2)[1,0])
print("Correlation coefficient dataset 3  ", np.corrcoef(x, y3)[1,0])
print("Correlation coefficient dataset 4  ", np.corrcoef(x4, y4)[1,0])
```

```python
# plot simple linear regression subplots using mathplotlib
# create regression line for data points and show on each plot 
# calculate and display slope and y intercept of regression lines

plt.subplot(221)
plt.scatter(x, y1)
plt.xlabel('X1')
plt.ylabel('Y1')
w, d = np.polyfit(x, y1, 1)
x_bf = np.array([3, 19])
y_bf = w * x_bf + d
plt.plot(x_bf, y_bf)
plt.axis([0, 20, 0, 15])
plt.text(2,12,'Dataset 1', fontsize = '15')
print ('slope & y intercept dataset 1', w, d)

plt.subplot(222)
plt.scatter(x, y2)
plt.xlabel('X2')
plt.ylabel('Y2')
w, d = np.polyfit(x, y2, 1)
x_bf = np.array([3, 19])
y_bf = w * x_bf + d
plt.plot(x_bf, y_bf)
plt.axis([0, 20, 0, 15])
plt.text(2,12,'Dataset 2', fontsize = '15')
print('slope & y intercept dataset 2', w, d)

plt.subplot(223)
plt.scatter(x, y3)
plt.xlabel('X3')
plt.ylabel('Y3')
w, d = np.polyfit(x, y3, 1)
x_bf = np.array([3, 19])
y_bf = w * x_bf + d
plt.plot(x_bf, y_bf)
plt.axis([0, 20, 0, 15])
plt.text(2,12,'Dataset 3', fontsize = '15')
print('slope & y intercept dataset 3', w, d)

plt.subplot(224)
plt.scatter(x4, y4)
plt.xlabel('X4')
plt.ylabel('Y4')
w, d = np.polyfit(x4, y4, 1)
x_bf = np.array([3, 19])
y_bf = w * x_bf + d
plt.plot(x_bf, y_bf)
plt.axis([0, 20, 0, 15])
plt.text(2,12,'Dataset 4', fontsize = '15')
print('slope & y intercept dataset 4', w, d)
plt.show()
```

## Section Five - Explain why it is interesting, referring to the plots and statistics from above


Upon reviewing the dataset summary statistics it would appear that they are almost identical sets of data.

All four datasets have identical (to two decimal places)  x and y means, standard deviation, and correlation coefficients. Looking at the descriptive statistics one could assume the data sets were almost identical and would have similar characteristics when plotted. That is far from the case as can been seen in the two sets of plots, above.

* Dataset 1 shows a linear relationship with points positioned around the regression line. This matches the prediction that could be made from the summary statistics.

* Dataset 2 is not at all linearwhen plotted.

* Dataset 3 does appear to be linear, however there is one outlier which is throwing the data off anf should ideally be excluded from the dataset for analysis. 

* Dataset 4 also contains an outlier and in this instance it is the outlier that is giving the impression of a linear relationship in the summay statistics. 

It really highlights that when you are looking at data you need to both visualise and analyse the data as it would be all too easy to make incorrect assumptions if relying on descriptive statistics alone. This dataset also highlights the affect that outliers can have on data. This also leads to the question of how to deal with outliers, whether to drop or not. An interesting article[5] by Karen Grace-Martin on this topic gave some easily understood scenarios where decisions to drop outliers or keep might arise.
