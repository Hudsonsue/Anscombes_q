# Anscome's quartet basic descriptive stats
- for each set of values of X and Y and using an for loop I have calculated and displayed the min/max and mean
- two


```python
print ("SUMMARY STATS")
print()
for i in range (0,8):
    feature = ['x1', 'x2', 'x3', 'x4', 'y1', 'y2', 'y3', 'y4']    
    print ("Minimum           ", feature [i], "is", np.min(A[:,i]))
    print ("Maximum           ", feature [i], "is",np.max(A[:,i]))
    print ("Mean              ", feature [i], "is", np.mean(A[:,i]))
    print ("Standard Deviation", feature [i], "is", np.std(A[:,i]))
    
    print()
```

```python
plt.subplot(221)
plt.scatter(x, y1)
plt.xlabel('X1')
plt.ylabel('Y1')
plt.axis([0, 20, 0, 15])
plt.text(2,12,'x1 vs y1', fontsize = '15')

plt.subplot(222)
plt.scatter(x, y2)
plt.xlabel('X2')
plt.ylabel('Y2')
plt.axis([0, 20, 0, 15])
plt.text(2,12,'x2 vs y2', fontsize = '15')

plt.subplot(223)
plt.scatter(x, y3)
plt.xlabel('X3')
plt.ylabel('Y3')
plt.axis([0, 20, 0, 15])
plt.text(2,12,'x3 vs y3', fontsize = '15')

plt.subplot(224)
plt.scatter(x4, y4)
plt.xlabel('X4')
plt.ylabel('Y4')
plt.axis([0, 20, 0, 15])
plt.text(2,12,'x4 vs y4', fontsize = '15')

plt.tight_layout
plt.show()
```
